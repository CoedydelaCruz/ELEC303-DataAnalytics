<h2>Fill Null Values with statistical values</h2>

```python

crimes['Location Description'] = crimes['Location Description'].fillna('Unknown')

# Fill numeric columns
crimes['Community Area'] = crimes['Community Area'].fillna(crimes['Community Area'].mode()[0])  # use most common value
crimes['X Coordinate'] = crimes['X Coordinate'].fillna(crimes['X Coordinate'].median())
crimes['Y Coordinate'] = crimes['Y Coordinate'].fillna(crimes['Y Coordinate'].median())
crimes['Latitude'] = crimes['Latitude'].fillna(crimes['Latitude'].median())
crimes['Longitude'] = crimes['Longitude'].fillna(crimes['Longitude'].median())

# For 'Location' (string with lat/long)
crimes['Location'] = crimes['Location'].fillna('(0,0)')

# Convert date columns
crimes['Date'] = pd.to_datetime(crimes['Date'], errors='coerce')
crimes['Updated On'] = pd.to_datetime(crimes['Updated On'], errors='coerce')

# Extract new time-based features
crimes['year'] = crimes['Date'].dt.year
crimes['month'] = crimes['Date'].dt.month
crimes['day'] = crimes['Date'].dt.day
crimes['dayofweek'] = crimes['Date'].dt.day_name()
crimes['hour'] = crimes['Date'].dt.hour

# Convert selected columns to categorical
categorical_cols = [
    'Primary Type', 'Description', 'Location Description',
    'Arrest', 'Domestic', 'District', 'Ward',
    'Community Area', 'FBI Code', 'year', 'month', 'dayofweek', 'hour'
]

crimes['Year'] = crimes['Date'].dt.year
crimes['Month'] = crimes['Date'].dt.month
crimes['DayOfWeek'] =crimes['Date'].dt.day_name()
crimes['Hour'] = crimes['Date'].dt.hour

for col in categorical_cols:
    crimes[col] = crimes[col].astype('category')
    
# Final check (should all be zero)
crimes.isnull().sum()
```

```python
fbi_arrest = crimes.groupby("FBI Code").agg(
    total=("ID","count"),
    arrests=("Arrest","sum")
).reset_index()
fbi_arrest["Arrest Rate %"] = (fbi_arrest["arrests"]/fbi_arrest["total"])*100

plt.figure(figsize=(12,6))
sns.barplot(x="FBI Code", y="Arrest Rate %", data=fbi_arrest)
plt.title("Arrest Rates by FBI Crime Code")
plt.xlabel("FBI Code")
plt.ylabel("Arrest Rate (%)")
plt.xticks(rotation=90)
plt.show()
```

<h2>INSIGHTS: Violent crime FBI codes, such as homicide and assault, lead to higher arrest rates. In contrast, property crime codes, like theft, show lower rates. This indicates different levels of enforcement effectiveness.</h2>

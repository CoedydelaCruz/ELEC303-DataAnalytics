<h1>7) Fill Null Values with statistical values</h1>

```python
accident['Accident_Severity'] = accident['Accident_Severity'].fillna(accident['Accident_Severity'].mode()[0])
accident['Light_Conditions'] = accident['Light_Conditions'].fillna(accident['Light_Conditions'].mode()[0])
accident['Road_Surface_Conditions'] = accident['Road_Surface_Conditions'].fillna(accident['Road_Surface_Conditions'].mode()[0])
accident['Number_of_Casualties'] = accident['Number_of_Casualties'].fillna(accident['Number_of_Casualties'].median())
accident['Number_of_Vehicles'] = accident['Number_of_Vehicles'].fillna(accident['Number_of_Vehicles'].median())
accident['District Area'] = accident['District Area'].fillna(accident['District Area'].mode()[0])
```

```python
import folium
from folium.plugins import HeatMap
import matplotlib.pyplot as plt
import seaborn as sns

# Filter data for Lambeth - only fatal accidents
Lambeth_df = accident[(accident["District Area"] == "Lambeth") & 
                      (accident["Accident_Severity"] == "Fatal")]

# Create HeatMap
locations = list(zip(Lambeth_df['Latitude'], Lambeth_df['Longitude']))
Lambeth_map = folium.Map(
    location=[Lambeth_df['Latitude'].mean(), Lambeth_df['Longitude'].mean()],
    zoom_start=12
)
HeatMap(locations).add_to(Lambeth_map)
Lambeth_map

```

<h1>Q26) How many fatal accidents occur in Lambeth across different days of the week, and are there noticeable peaks?</h1>

```python
plt.figure(figsize=(8,5))
sns.countplot(data=Lambeth_df, x='DayOfWeek', palette='Reds')
plt.title("Fatal Accidents by Day of Week in Lambeth")
plt.xticks(ticks=range(7), labels=['Mon','Tue','Wed','Thu','Fri','Sat','Sun'])
plt.tight_layout()
plt.show()
```

<h3>Insights:This shows that fatal accidents in Lambeth are happen when weekends approach, especially Saturday nights, and these events are caused by nightlife, alcohol consumption, and increased pedestrian movement in areas where they are crowded with many people. Accidents during the weeK between Mondays and Fridays are fewer in numbers. This basically reflects very light regular commuting particularly in risky late-night existence and activities.</h3>
